```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
from datetime import datetime

# Plotting
plt.figure(figsize=(16, 6), dpi=300) # increasing the reolutiion for the badge

# Load the image
img = plt.imread(r'C:\Users\smorr\Downloads\Arsenal_Badge2.png')

# Plot the line
plt.plot(date_strings, values, color='black', linestyle='-')

#dates for title - extracting first and last dates
first_date = date_strings[0]
last_date = date_strings[-1]

# Function to format date string
def format_date(date_string):
    # Convert date string to datetime object
    date_object = datetime.strptime(date_string, '%m-%Y')
    # Format datetime object to 'Month Year' format
    formatted_date = date_object.strftime('%B %Y')
    return formatted_date

first_date = format_date(first_date) # Output: September 2019
last_date = format_date(last_date) # Output: March 2024

# Plot the data
for date, value in zip(date_strings, values): 
    # Using the Arsenal badge as a custom marker
    imagebox = OffsetImage(img, zoom=0.2)  # Adjust zoom to scale the image
    ab = AnnotationBbox(imagebox, (date, value), frameon=False)
    plt.gca().add_artist(ab)
    
plt.title(f'Bukayo Saka - Market Value Over Time\n({first_date} to {last_date})', fontsize=16, fontweight='bold' )
plt.xlabel('Date', fontsize=14)
plt.ylabel('Market Value (in million euros)', fontsize=14)
plt.xticks(rotation=0)
plt.grid(False)
plt.tight_layout()
plt.show()
```
